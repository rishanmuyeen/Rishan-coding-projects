```python
#dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import difflib  #to get the input
from sklearn.feature_extraction.text import TfidfVectorizer  #to convert characters to meaning full numerical values
from sklearn.metrics.pairwise import cosine_similarity  #gives a similarity score to recommed 

```

```python
#data collection
dataframe = pd.read_csv('movies.csv')
dataframe.head()
```

```python
dataframe.shape
```

```python
selected_features = ['genres','runtime','title','vote_count','cast','director']
print(selected_features)
```

```python
#replacing null values
for i in selected_features:
    dataframe[i] = dataframe[i].fillna('')
```

```python
combined_features = dataframe['genres'] + ' ' + dataframe['runtime'].astype(str) + ' ' + dataframe['title'] + ' ' + dataframe['vote_count'].astype(str) + ' ' + dataframe['cast'] + ' ' + dataframe['director']
```

```python
combined_features
```

```python
#converting the characters to numerical value 
feature_extration = TfidfVectorizer()
extracted_combined_features = feature_extration.fit_transform(combined_features)
```

```python
print(extracted_combined_features)
```

```python
similarity = cosine_similarity(extracted_combined_features)
similarity.shape
```

```python
#getting user input
#test sample
user_input = str(input('Enter your favourite movie name: '))
```

```python
#creating a list that contains all the movie names
movie_names = dataframe['title'].tolist()
movie_names
```

```python
#finding the close matach to our favourite movie
match = difflib.get_close_matches(user_input,movie_names)
match
```

```python
#getting the first index
close_match = match[0]
close_match
```

```python
#getting the title of it
index_close_match = dataframe[dataframe.title == close_match]['index'].values[0]
index_close_match
```

```python
#using the similarity cosign similarity to check with other movies
similarity_score = list(enumerate(similarity[index_close_match]))
similarity_score
#from 0.0-1.0 1 is the highest match
```

```python
#sorting the list in decending order so we can get closest related movie 
sorted_similarity_score = sorted(similarity_score, key= lambda x:x[1], reverse=True)
sorted_similarity_score
```

```python
print('Here are recommended movies :')
i = 1
for movie in sorted_similarity_score:
    index = movie[0]
    title_index = dataframe[dataframe.index == index]['title'].values[0]
    if(i<31):
        print(i, '. ' , title_index)
        i+=1
        
```

```python
#this is the system for the movie recommendation
#getting the input
user_input = str(input('Enter your favourite movie name: '))
#getting all the movie names from the dataframe
movie_names = dataframe['title'].tolist()
#using difflib function to fund the closest similarity
match = difflib.get_close_matches(user_input,movie_names)
#getting the first value
close_match = match[0]
#getting the index value from the dataframe
index_close_match = dataframe[dataframe.title == close_match]['index'].values[0
#finding the closed values
#enumerate is a function that carried a operation to all the elements
similarity_score = list(enumerate(similarity[index_close_match]))
#sorting them in decending order
sorted_similarity_score = sorted(similarity_score, key= lambda x:x[1], reverse=True)

print('Here are recommended movies :')
i = 1
for movie in sorted_similarity_score:
    index = movie[0]
    row = dataframe.iloc[index]
    title = row['title']
    director = row['director']
    if(i<31):
        print(i, '. ' , title, '  directed by ', director)
        i+=1
        
```
